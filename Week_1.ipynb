**Variables in Python**

Variables in Python are used to store data that you want to use later in your code. Think of a variable as a labeled box where you can store a value. When you assign a value to a variable, Python remembers that value and allows you to use it later by referring to the variable's name.

**1. Assigning Values to Variables**

You can assign a value to a variable using the = operator. The value on the right side of the = is stored in the variable on the left side.

```python
# Example of variable assignment
gene_count = 42
organism_name = "E. coli"
growth_rate = 1.23
is_bacteria = True
```

- gene_count stores an integer value 42.
- organism_name stores a string value "E. coli".
- growth_rate stores a floating-point number 1.23.
- is_bacteria stores a boolean value True.

**2. Variable Naming Conventions**

When naming variables in Python, there are some important rules and conventions to follow:

- Variable names should start with a letter (a-z, A-Z) or an underscore (_).
- The name can contain letters, numbers (0-9), and underscores, but cannot start with a number.
- Variable names are case-sensitive, so gene_count and Gene_Count would be considered different variables.
- It's a good practice to use descriptive names that clearly indicate what the variable represents.

```python
# Examples of valid variable names
dna_sequence = "ATGCGT"
temperature_celsius = 37.0

# Examples of invalid variable names
# 1gene_count = 42      # Invalid: starts with a number
# gene-count = 42       # Invalid: hyphen is not allowed
```

```python
# Printing variables
print("Gene Count:", gene_count)
print("Organism Name:", organism_name)
print("Growth Rate:", growth_rate)
print("Is Bacteria:", is_bacteria)
```

**Data Types**

In Python, every value has a specific data type. Data types define the nature of the data and determine what kind of operations can be performed on that data. Understanding data types is fundamental to programming in Python.

**1. Basic Data Types**

Let's start by looking at some of the basic data types that you'll frequently encounter in Python: integers, floats, strings, and booleans.

**a) Integer (int)**

Integers are whole numbers without a decimal point. They can be positive or negative.

```python
# Example of an integer
int_var = 10
negative_int = -5

print("Integer:", int_var)
print("Negative Integer:", negative_int)
```

You can perform various arithmetic operations with integers, such as addition, subtraction, multiplication, and division.

```python
# Arithmetic with integers
sum_result = int_var + negative_int
product_result = int_var * 2

print("Sum:", sum_result)
print("Product:", product_result)
```

**b) Floating-Point Number (float)**

Floats represent real numbers and include a decimal point. They are used when you need more precision than integers provide.

```python
# Example of a float
float_var = 3.14
large_float = 1.23e5  # Scientific notation for 1.23 * 10^5

print("Float:", float_var)
print("Large Float:", large_float)
```

Floats are especially useful in scientific calculations, where precision is critical.

```python
# Example of float operations
area_of_circle = 3.14 * (2 ** 2)  # Area = πr^2 for a circle with radius 2
print("Area of Circle:", area_of_circle)
```

**c) String (str)**

Strings are sequences of characters enclosed in either single quotes (') or double quotes ("). Strings can include letters, numbers, spaces, and special characters.

```python
# Example of a string
string_var = "DNA sequence"
sequence_description = "This sequence is of gene X."

print("String:", string_var)
print("Sequence Description:", sequence_description)
```

Strings can be concatenated (joined together) and repeated using the + and * operators, respectively.

```python
# String concatenation and repetition
full_message = string_var + " analysis."
repeated_message = "A" * 5  # Repeats 'A' five times

print("Full Message:", full_message)
print("Repeated Message:", repeated_message)
```

**d) Boolean (bool)**

Booleans represent one of two possible values: True or False. They are often used in conditional statements to control the flow of a program.

```python
# Example of a boolean
bool_var = True
is_valid = False

print("Boolean (True):", bool_var)
print("Boolean (False):", is_valid)
```

Booleans are commonly used in comparisons and logical operations.

```python
# Boolean expressions
is_greater = int_var > 5
is_equal = string_var == "DNA sequence"

print("Is Greater:", is_greater)
print("Is Equal:", is_equal)
```

**4. Advanced Data Types (Brief Overview)**

Python also supports more complex data types, such as:

- Lists: Ordered collections of items, which can be of mixed types.
- Tuples: Immutable ordered collections of items.
- Dictionaries: Collections of key-value pairs.
- Sets: Unordered collections of unique items.

In addition to basic data types like integers, floats, strings, and booleans, Python provides several more complex data types that are essential for organizing and manipulating collections of data. These include lists, tuples, dictionaries, and sets. Each of these data types serves a specific purpose and has unique characteristics.

**1. Lists**

Lists are ordered collections of items, which means the items have a defined sequence, and you can access them by their position (index) in the list. Lists are mutable, meaning you can change their contents after they are created. Lists can contain items of different data types, including other lists.

**a) Creating a List**

You can create a list by placing a comma-separated sequence of items inside square brackets [].

```python
# Creating a list
gene_list = ["BRCA1", "TP53", "MYC", "EGFR"]
mixed_list = [42, 3.14, "Gene", True]

print("Gene List:", gene_list)
print("Mixed List:", mixed_list)
```

**b) Accessing List Elements**

You can access individual elements in a list by their index. Python uses zero-based indexing, so the first element has an index of 0.

```python
# Accessing elements
first_gene = gene_list[0]
second_gene = gene_list[1]

print("First Gene:", first_gene)
print("Second Gene:", second_gene)
```

**c) Modifying Lists**

Since lists are mutable, you can change their contents, add new elements, or remove elements.

```python
# Modifying a list
gene_list[1] = "PTEN"  # Changing the second gene
gene_list.append("CDH1")  # Adding a new gene at the end
gene_list.remove("MYC")  # Removing a gene by value

print("Modified Gene List:", gene_list)
```

**d) List Operations**

Lists support various operations, such as slicing, concatenation, and repetition.

```python
# List slicing
subset_genes = gene_list[1:3]  # Get genes from index 1 to 2
print("Subset of Genes:", subset_genes)

# Concatenation
combined_list = gene_list + ["ALK", "BRAF"]
print("Combined List:", combined_list)

# Repetition
repeated_list = gene_list * 2
print("Repeated List:", repeated_list)
```

**2. Tuples**

Tuples are similar to lists in that they are ordered collections of items. However, tuples are immutable, meaning once you create a tuple, you cannot change its contents. This immutability makes tuples useful for storing data that should not be modified.

**a) Creating a Tuple**

You can create a tuple by placing a comma-separated sequence of items inside parentheses ().

```python
# Creating a tuple
gene_tuple = ("BRCA1", "TP53", "MYC")
single_element_tuple = ("BRCA1",)  # Note the comma, necessary for single element tuples

print("Gene Tuple:", gene_tuple)
print("Single Element Tuple:", single_element_tuple)
```

**b) Accessing Tuple Elements**

Just like lists, you can access elements in a tuple by their index.

```python
# Accessing elements
first_gene = gene_tuple[0]
print("First Gene in Tuple:", first_gene)
```

**c) Tuple Operations**

Tuples support many of the same operations as lists, such as slicing, concatenation, and repetition. However, since tuples are immutable, you cannot modify them (e.g., you can't append or remove elements).

```python
# Tuple slicing
subset_genes = gene_tuple[0:2]
print("Subset of Tuple:", subset_genes)

# Concatenation
new_tuple = gene_tuple + ("EGFR",)
print("Concatenated Tuple:", new_tuple)

# Repetition
repeated_tuple = gene_tuple * 2
print("Repeated Tuple:", repeated_tuple)
```

While you can concatenate a new element to a tuple, this operation actually creates a new tuple, and the original tuple remains unchanged.


```python
# Creating a tuple
gene_tuple = ("BRCA1", "TP53", "MYC")
print("Original Gene Tuple:", gene_tuple)

# Trying to remove an element from the tuple
try:
    del gene_tuple[1]  # Attempting to delete the element at index 1 ("TP53")
except TypeError as e:
    print("Error:", e)  # This will raise a TypeError because tuples are immutable

# The original tuple remains unchanged
print("Gene Tuple after failed removal attempt:", gene_tuple)
```

Explanation:

- The del gene_tuple[1] statement tries to delete the element at index 1, which is "TP53". Since tuples are immutable, this will raise a TypeError.

- The original tuple remains unchanged after the failed removal attempt.

**3. Dictionaries**

Dictionaries are collections of key-value pairs, where each key is unique, and each key is associated with a value. Dictionaries are unordered, meaning the items do not have a defined sequence. They are extremely useful for storing and retrieving data based on a key.

**a) Creating a Dictionary**

You can create a dictionary by placing a comma-separated sequence of key-value pairs inside curly braces {}. The key and value are separated by a colon :

```python
# Creating a dictionary
gene_dict = {
    "BRCA1": "Breast cancer susceptibility gene",
    "TP53": "Tumor protein p53",
    "MYC": "Oncogene associated with many cancers"
}

print("Gene Dictionary:", gene_dict)
```

**b) Accessing Dictionary Values**

You can access the value associated with a specific key using square brackets [].

```python
# Accessing a value
brca1_description = gene_dict["BRCA1"]
print("BRCA1 Description:", brca1_description)
```

**c) Modifying Dictionaries**

Dictionaries are mutable, so you can change the value associated with a key, add new key-value pairs, or remove existing ones.

```python
# Modifying a dictionary
gene_dict["BRCA1"] = "Updated description for BRCA1"  # Update value
gene_dict["EGFR"] = "Epidermal growth factor receptor"  # Add a new key-value pair
del gene_dict["MYC"]  # Remove a key-value pair

print("Updated Gene Dictionary:", gene_dict)
```

**d) Dictionary Operations**

Dictionaries support various operations, including checking if a key exists, iterating over keys and values, and retrieving all keys or values.

```python
# Check if a key exists
has_tp53 = "TP53" in gene_dict
print("Contains TP53:", has_tp53)

# Iterating over keys and values
for gene, description in gene_dict.items():
    print(f"Gene: {gene}, Description: {description}")

# Retrieving all keys or values
keys = gene_dict.keys()
values = gene_dict.values()
print("Keys:", keys)
print("Values:", values)
```

**4. Sets**

Sets are unordered collections of unique items. Sets are useful when you want to store items but ensure that there are no duplicates. They also support operations like unions, intersections, and differences, which are useful for tasks involving sets of data.

**a) Creating a Set**

You can create a set by placing a comma-separated sequence of items inside curly braces {} or by using the set() function.

```python
# Creating a set
gene_set = {"BRCA1", "TP53", "MYC", "BRCA1"}  # Note: Duplicates will be removed

print("Gene Set:", gene_set)
```

**b) Modifying Sets**

Sets are mutable, so you can add or remove elements, but since they are unordered, you cannot access items by index.

```python
# Modifying a set
gene_set.add("EGFR")  # Add a new element
gene_set.discard("MYC")  # Remove an element

print("Modified Gene Set:", gene_set)
```

**c) Set Operations**

Sets support various mathematical set operations, such as union, intersection, and difference.

```python
# Set operations
another_set = {"EGFR", "ALK"}

# Union: Combine two sets
union_set = gene_set.union(another_set)
print("Union:", union_set)

# Intersection: Find common elements
intersection_set = gene_set.intersection(another_set)
print("Intersection:", intersection_set)

# Difference: Find elements in gene_set but not in another_set
difference_set = gene_set.difference(another_set)
print("Difference:", difference_set)
```

**Basic Operators**

Python supports various operators for arithmetic, comparison, and logical operations.

```python
# Arithmetic Operators
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)

# Comparison Operators
print("Equal:", a == b)
print("Not Equal:", a != b)
print("Greater Than:", a > b)
print("Less Than:", a < b)

# Logical Operators
print("And:", a > 5 and b < 5)
print("Or:", a > 5 or b > 5)
print("Not:", not(a > 5))
```

#Writing and Executing Simple Python Scripts in Colab

In this section, we will write a simple Python script that calculates the GC content of a DNA sequence. The GC content is the percentage of nucleotides in a DNA sequence that are either guanine (G) or cytosine (C). This is an important metric in molecular biology as it can influence the stability of DNA and its behavior in different environments.

```python
# Calculate GC Content of a DNA sequence

# Sample DNA sequence
dna_sequence = "ATGCGCGATCGTACGCTAGC"

# Count G and C nucleotides
g_count = dna_sequence.count('G')
c_count = dna_sequence.count('C')

# Calculate GC content
gc_content = (g_count + c_count) / len(dna_sequence) * 100

print("DNA Sequence:", dna_sequence)
print("GC Content: {:.2f}%".format(gc_content))
```

**Explanation:**

- We define a DNA sequence.
- We count the occurrences of 'G' and 'C'.
- We calculate the GC content as the percentage of 'G' and 'C' in the sequence.

#Conditionals (if, elif, else)

**Overview:**

Conditionals are fundamental in programming and are used to control the flow of a program. They allow you to execute specific blocks of code only when certain conditions are met. This is particularly useful when you want your program to make decisions and act differently based on varying inputs or states.

**Syntax:**

if condition:
    # Code block executed if condition is True
elif another_condition:
    # Code block executed if another_condition is True
else:
    # Code block executed if all conditions are False
    

- **if Statement:** Checks the first condition. If it’s True, the associated code block runs.
- **elif Statement:** Short for "else if," this checks another condition if the previous if or elif statements are False.
- **else Statement:** This block runs if none of the previous conditions are True. It’s optional but useful as a fallback.

**Example:** Determining if a Number is Positive, Negative, or Zero

```python
# Example: Determining if a number is positive, negative, or zero
number = 10

if number > 0:
    print("The number is positive.")
elif number < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

**Explanation:**

- If number > 0: This checks if the number is greater than zero. If true, it prints "The number is positive."
- Elif number < 0: If the first condition is false, this checks if the number is less than zero. If true, it prints "The number is negative."
- Else: If neither of the above conditions is true, it prints "The number is zero."

**Additional Examples:**

1. Simple Voting Eligibility Check:

```python
age = 18

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

**Explanation:**

- If the age is 18 or older, the message says you can vote. Otherwise, it indicates you're not eligible.

2. Nested Conditionals: You can also nest conditionals within each other to check multiple related conditions.

```python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    if score >= 85:
        print("Grade: B+")
    else:
        print("Grade: B")
else:
    print("Grade: C")
```

**Explanation:**

- The code first checks if the score is 90 or above for an 'A'.
- If not, it checks if the score is at least 80. If the score is also 85 or higher, it assigns a 'B+', otherwise, a 'B'.
- If none of these conditions are met, it assigns a 'C'.

**a) For Loops**

**Overview:**

For loops in Python are used to iterate over a sequence (like a list, tuple, string, or range of numbers) and execute a block of code for each item in that sequence. They are particularly useful when you want to perform the same operation on each item in a collection of data.

**Syntax:**

for item in sequence:
    # Code block to be executed for each item

- item: A temporary variable that takes the value of the current element in the sequence during each iteration.
- sequence: The collection of elements you want to loop through (e.g., a list, string, or range).

**Example:**

Let's take a closer look at a for loop that iterates over a list of gene names.

```python
# Example: Iterating over a list of genes
genes = ["BRCA1", "TP53", "MYC", "EGFR"]

for gene in genes:
    print(gene)
```

**Explanation:**

- Iteration: The loop starts with the first item in the genes list ("BRCA1") and assigns it to the variable gene. The code block inside the loop then prints "BRCA1".
- Next Iteration: The loop moves to the next item in the list ("TP53"), assigns it to gene, and prints it. This process repeats for each item in the list.
- Completion: The loop continues until all items in the list have been processed.

**Additional Example:** Iterating Over a String

You can also use a for loop to iterate over each character in a string.

```python
# Example: Iterating over a DNA sequence string
dna_sequence = "ATGCGC"

for nucleotide in dna_sequence:
    print(nucleotide)
```

**Explanation:**

Here, nucleotide takes each character in the dna_sequence string, and the loop prints each nucleotide one by one.

**Exercise:**

Write a Python script that iterates over a list of DNA sequences and prints the GC content of each sequence.

```python
sequences = ["ATGCGCGATCGTACGCTAGC", "ATATATATAT", "GGGGCCCC"]

for seq in sequences:
    g_count = seq.count('G')
    c_count = seq.count('C')
    gc_content = (g_count + c_count) / len(seq) * 100
    print(f"Sequence: {seq}, GC Content: {gc_content:.2f}%")
```

**Exercise Explanation:**

- Iterating Over Sequences: The loop iterates over each DNA sequence in the sequences list.
- GC Content Calculation: For each sequence, the loop calculates the GC content and prints it.

**b) While Loops**

**Overview:**

While loops in Python are used to repeat a block of code as long as a specific condition is True. Unlike for loops, which iterate over a sequence, while loops continue to execute as long as the condition remains true, making them useful for situations where the number of iterations isn't predetermined.

**Syntax:**

while condition:
    # Code block to be executed while condition is True

- condition: A boolean expression that is evaluated before each iteration. If it’s True, the code block runs; if it’s False, the loop stops.

**Example:** Counting Down from 5 to 1

```python
# Example: Counting down from 5 to 1
count = 5

while count > 0:
    print(count)
    count -= 1  # Decrease count by 1
```

**Explanation:**

- Initial Condition: The loop begins by checking if count is greater than 0. Since count is initially 5, the condition is true, so the loop enters its code block.
- Decrementing: Inside the loop, count is printed, and then it’s decreased by 1 using count -= 1.
- Reevaluation: After each iteration, the condition (count > 0) is re-evaluated. The loop continues until count reaches 0.

**Additional Example:** Finding a Target Value

Here's an example where we use a while loop to find the first occurrence of a target value in a list.

```python
# Example: Finding the first occurrence of a target value
numbers = [10, 20, 30, 40, 50]
target = 30
index = 0

while index < len(numbers) and numbers[index] != target:
    index += 1

if index < len(numbers):
    print(f"Found {target} at index {index}")
else:
    print(f"{target} not found in the list.")
```

**Explanation:**

- Condition: The loop continues as long as index is less than the length of the list and the current number isn't equal to the target.
- Finding the Target: When the target value is found, the loop stops, and the index of the target is printed.

**Exercise:**

Write a Python script that finds the first occurrence of a stop codon (TAA, TAG, TGA) in a DNA sequence using a while loop.

```python
dna_sequence = "ATGCGCGATCGTAACTAGCTAGGCGTGA"

stop_codons = ["TAA", "TAG", "TGA"]
position = 0
found = False

while position <= len(dna_sequence) - 3 and not found:
    codon = dna_sequence[position:position+3]
    if codon in stop_codons:
        print(f"Stop codon {codon} found at position {position}")
        found = True
    position += 3

if not found:
    print("No stop codon found.")
```

**Exercise Explanation:**

- Looping Through the Sequence: The loop iterates over the DNA sequence in chunks of three nucleotides (a codon) at a time.
- Stop Codon Detection: When a stop codon is found, the loop prints its position and stops further iterations.
